```python
def init_transforms(data):
    data["City"] = data["City"].apply(lambda x:str(x))
    data["is_big_city"] = data["City Group"].apply(lambda x:int(x=="Big Cities"))
    data["start_year"] = data["Open Date"].apply(lambda x:int(x[6:]))
    data["start_month"] = data["Open Date"].apply(lambda x:int(x[:2]))
    data["start_day"] = data["Open Date"].apply(lambda x:int(x[3:5]))
    data["num_years"] = data["start_year"].apply(lambda x:2015 - x)
    data["num_months"] = data["num_years"].apply(lambda x:x*12) - data["start_month"].apply(lambda x:12-x)
    data["Type_FC"] = data["Type"].apply(lambda x:int(x=="FC"))
    data["Type_IL"] = data["Type"].apply(lambda x:int(x=="IL"))
    data["Type_DT"] = data["Type"].apply(lambda x:int(x=="DT"))
    data["Type_MB"] = data["Type"].apply(lambda x:int(x=="MB"))
    data = data.drop(["Id","City","City Group","Open Date",
                      "Type","start_year","start_month","start_day","num_years"],axis=1)
    return data
```

```python
def load_data(data_type,shuffle=False):
    assert(data_type == "train" or data_type == "test" or data_type == "cv")
    if data_type == "cv":   data_type = "train"
    path = os.path.join(load_params().data_dir,data_type + ".csv")
    data = read_csv(path)
    # print data.columns.values
    if data_type != "test":
        if shuffle: data = data.reindex(np.random.permutation(data.index))
        #y = data.revenue.values
        #data = data.drop(["revenue"],axis=1)
        data = init_transforms(data)
        return data
        #X = data.as_matrix()
        #return np.log(X + 1),y
    else:
        data = init_transforms(data)
        return data
        #X = data.as_matrix()
        #return np.log(X + 1)
```

```python
from sklearn.preprocessing import StandardScaler
X = load_data("train")
# Applying transforms here
cols =  X.columns.values.tolist()
cols.remove('revenue')
cols = ['P8' ,'P10' ,'P12' ,'P15' ,'P17' ,'P18' ,'P22' ,'P28' ,'P32' ,'num_months']
print "Applied log(X+1)"
for vary in cols:
    X[vary] = X[[vary]].apply(lambda x:np.log(x + 1)) 
    print ggplot(aes(x='revenue',y=vary), data=X) + geom_point()
```

```python
X = load_data("train")
# Applying transforms here
cols =  X.columns.values.tolist()
cols.remove('revenue')
from sklearn.feature_selection import RFECV
from sklearn.linear_model import Ridge,LogisticRegression
y = X["revenue"]
print y
X = X.drop(["revenue"],axis=1)
X = np.log(X.as_matrix() + 1)
print "Linear Regressors"
clf_lr = RFECV(Ridge(),scoring="mean_squared_error",cv=137)
clf_lr.fit(X,np.log(y))
cols_lr = np.array(cols)[clf_lr.support_].tolist()
print "Columns for Linear Regression"
print cols_lr
print "Remaining"
print list(set(cols) - set(cols_lr))
print "Logistic Regression"
clf_lr = RFECV(LogisticRegression(),scoring="mean_squared_error",cv=137)
clf_lr.fit(X,np.log(y))
cols_lr = np.array(cols)[clf_lr.support_].tolist()
print "Columns for Logistic Regression"
print cols_lr
print "Remaining"
print list(set(cols) - set(cols_lr))
```
